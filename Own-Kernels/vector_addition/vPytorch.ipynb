```python
import torch
```

```python
print(torch.cuda.is_available())
print(torch.__version__)
```

```python
N = 2**25
A = torch.ones(N, dtype=torch.float32, device="cpu")
B = 2.0 * torch.ones(N, dtype=torch.float32, device="cpu")
d_A = A.to("cuda")
d_B = B.to("cuda")
```

Kernel only

```python
%%timeit
_ = d_A + d_B
torch.cuda.synchronize()
```

Kernel inclusive host <-> device copy operations

```python
%%timeit
d_A = A.to("cuda")
d_B = B.to("cuda")
d_C = d_A + d_B
d_C.to("cpu")
torch.cuda.synchronize()
```
